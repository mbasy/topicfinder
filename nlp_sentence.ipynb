```python
import pandas as pd
import numpy as np
import re
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# load the CSV file into a pandas DataFrame
data = pd.read_excel('data.xlsx')

# remove punctuation and digits from sentences
def clean_text(text):
    text = text.translate(str.maketrans('', '', string.punctuation + string.digits))
    text = re.sub(r'\s+', ' ', text).strip()
    return text

data['clean_text'] = data['Ärende'].apply(clean_text)

# vectorize the sentences using TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(data['clean_text'])

# perform KMeans clustering on the sentence vectors

k = 10 # number of clusters
#kmeans = KMeans(n_clusters=k, random_state=42)
kmeans = KMeans(n_clusters=k, init='k-means++', max_iter=200, n_init=1, random_state=42)
kmeans.fit(X)


# assign cluster labels to the sentences
data['cluster_label'] = kmeans.labels_

# print the sentences in each cluster
for i in range(k):
    cluster_i = data[data['cluster_label'] == i]
    print(f'Cluster {i}:')
    for text in cluster_i['Ärende']:
        print(f'\t{text}')
```
